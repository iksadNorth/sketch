```python
file_name = 'confidential/seoul-covid19_6_30_.csv'
df = pd.read_csv(file_name)
```

```python
df = df.sort_values(by='연번', ascending=False)
df.head()
```

```python
pd.Series([1,3,-4,5,-6]).plot.bar(title='한글 제목')
plt.axhline(0, color='black')
```

```python
df['확진일자'] = pd.to_datetime('2020-' + df['확진일'].str.replace('.', '-'))
df['월'] = df['확진일자'].dt.month
df['주'] = df['확진일자'].dt.week

df[['확진일', '확진일자', '월', '주']].head()
```

```python
df['확진일자'].value_counts().plot(figsize=(15,4))
```

```python
df['월일'] = df['확진일자'].astype(str).map(lambda x : x[-5:])
day_count = df['월일'].value_counts().sort_index()
graph = day_count.plot(figsize=(15,4))
for idx, i in enumerate(day_count):
    if i >= 20:
        graph.text(x=idx, y=i, s=i)
```
